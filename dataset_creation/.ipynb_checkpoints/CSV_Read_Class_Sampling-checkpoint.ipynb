```python
def preprocess_category_names(c_name):
    
    # lowercase
    c_name = c_name.lower()
    
    # tokenize
    c_name = nltk.word_tokenize(c_name)
    
    c_final =  [word for word in c_name if word not in stopwords.words('english')]
    
    return c_final
```

```python
def embedding_extractor(w_list):
    
    final_list = [0] * 300
    
    for w in w_list:
        final_list += fasttext[w]
        
    # normalize
    final_list = final_list / len(w_list)
    
    return final_list.reshape(1, -1)
```

```python
main_categories_embeddings = []
sub_categories_embeddings = []
```

```python
for i in range(len(main_categories)):
    main_categories_embeddings.append(embedding_extractor(preprocess_category_names(main_categories[i])))
```

```python
for i in range(len(sub_categories)):
    temp = []
    for j in range(len(sub_categories[i])):
        temp.append(embedding_extractor(preprocess_category_names(sub_categories[i][j])))
    sub_categories_embeddings.append(temp)
```

```python
main_similarity_matrix = np.zeros((len(main_categories_embeddings), len(main_categories_embeddings)))

for i in range(len(main_categories_embeddings)):
    for j in range(len(main_categories_embeddings)):
        main_similarity_matrix[i,j] = cosine_similarity(main_categories_embeddings[i], main_categories_embeddings[j])
```

```python
plt.figure(figsize=(50,50))
plt.matshow(main_similarity_matrix,aspect='auto')
plt.show();
```

```python
most_2_less = []
most_2_less_p = []
```

```python
for i in range(len(main_categories)):
    list1, list2 = zip(*sorted(zip(main_similarity_matrix[i,:], list(main_categories)), reverse=True))
    most_2_less.append(list2)
    most_2_less_p.append(list1)
```

```python
for i in range(len(most_2_less)):
    for j in range(len(most_2_less[i])):
        print("{} -- {:.2f}".format(most_2_less[i][j], most_2_less_p[i][j]))
    
    print("\n\n")
    print("----")
    print("\n\n")
```

```python
## K-Means to construct 3-classes with using their embedding matrix

# drop nan class
main_categories_kmeans = list(main_categories)[:-1]
main_categories_embeddings_kmeans = list(main_categories_embeddings)[:-1]

main_categories_embeddings_kmeans = np.array(main_categories_embeddings_kmeans).reshape(19,300)
```

```python
main_categories_kmeans
```

```python
from sklearn.cluster import KMeans
from nltk.cluster import KMeansClusterer

kmeans = KMeansClusterer(3, distance=nltk.cluster.util.euclidean_distance, avoid_empty_clusters=True, repeats=100)
kmeans_labels = kmeans.cluster(main_categories_embeddings_kmeans, assign_clusters=True)
```

```python
group_1 = []
group_2 = []
group_3 = []

for i in range(len(kmeans_labels)):
    if kmeans_labels[i] == 0:
        group_1.append(main_categories[i])
    elif kmeans_labels[i] == 1:
        group_2.append(main_categories[i])
    elif kmeans_labels[i] == 2:
        group_3.append(main_categories[i])
```

## Download File Preparation

```python
train_classes = ['Family Life', 'Hobbies and Crafts', 'Holidays and Traditions', 'Home and Garden', 'Sports and Fitness', 'Education and Communications', 'Pets and Animals', 'Philosophy and Religion', 'Relationships', 'Work World', 'Youth']
validation_classes = ['Arts and Entertainment', 'Cars & Other Vehicles', 'Computers and Electronics','Finance and Business']
test_classes = ['Food and Entertaining', 'Health', 'Personal Care and Style','Travel']
```

```python
# get subclasses
df.columns
```

```python
train_subcategories = []
val_subcategories = []
test_subcategories = []

for elem in train_classes:
    train_subcategories += df.category_2[df["category_1"] == elem].dropna().unique().tolist()

for elem in validation_classes:
    val_subcategories += df.category_2[df["category_1"] == elem].dropna().unique().tolist()
    
for elem in test_classes:
    test_subcategories += df.category_2[df["category_1"] == elem].dropna().unique().tolist()
```

```python
len(train_subcategories),len(val_subcategories),len(test_subcategories)
```

```python
# select: 64 for training
#         12 for validation
#         24 for testing

# Not able to random select because some classes do NOT have >= 100 videos
train_classes = []
validation_classes = []
test_classes = []

for elem in train_subcategories:
    n_v = len(df.video_id[df["category_2"] == elem].dropna().unique().tolist())
    
    if n_v >= 100:
        train_classes.append(elem)
        
for elem in val_subcategories:
    n_v = len(df.video_id[df["category_2"] == elem].dropna().unique().tolist())
    
    if n_v >= 100:
        validation_classes.append(elem)
        
for elem in test_subcategories:
    n_v = len(df.video_id[df["category_2"] == elem].dropna().unique().tolist())
    
    if n_v >= 100:
        test_classes.append(elem)
```

```python
len(train_classes),len(validation_classes),len(test_classes)
```

```python
train_classes = random.sample(train_classes, 64)
validation_classes = random.sample(validation_classes, 12)
test_classes = random.sample(test_classes, 24)
```

```python
len(train_classes),len(validation_classes),len(test_classes)
```

```python
# get video id's for each class

train_v_ids = []
val_v_ids = []
test_v_ids = []

for elem in train_classes:
    
    class_all_v_ids = df.video_id[df["category_2"] == elem].dropna().unique().tolist()
    v_ids = random.sample(class_all_v_ids, 100)
    
    train_v_ids.append(v_ids)
    
for elem in validation_classes:
    
    class_all_v_ids = df.video_id[df["category_2"] == elem].dropna().unique().tolist()
    v_ids = random.sample(class_all_v_ids, 100)
    
    val_v_ids.append(v_ids)

for elem in test_classes:
    
    class_all_v_ids = df.video_id[df["category_2"] == elem].dropna().unique().tolist()
    v_ids = random.sample(class_all_v_ids, 100)
    
    test_v_ids.append(v_ids) 
```

```python
np.array(train_v_ids).shape, np.array(val_v_ids).shape, np.array(test_v_ids).shape
```

```python
# get the internal full links of v_ids
train_v_link = []
val_v_link = []
test_v_link = []

full_link_txt_path = os.path.join("../../other", "download_instructions")
full_link_txt_path = os.path.join(full_link_txt_path, "howto100m_videos.txt")

file1 = open(full_link_txt_path, 'r')
lines = file1.readlines()

# remove the \n from each line
for i in range(len(lines)):
    lines[i] = lines[i].strip()
    
line_beginner = "http://howto100m.inria.fr/dataset/"
```

```python
for v_ids in train_v_ids:
    temp = []
    
    for v_id in v_ids:
        
        try:
            l_idx = lines.index(line_beginner+v_id+".mp4")
            temp.append(line_beginner+v_id+".mp4")
                
        except:
            temp.append(line_beginner+v_id+".webm")
    
    train_v_link.append(temp)
```

```python
for v_ids in val_v_ids:
    temp = []
    
    for v_id in v_ids:
        
        try:
            l_idx = lines.index(line_beginner+v_id+".mp4")
            temp.append(line_beginner+v_id+".mp4")
                
        except:
            temp.append(line_beginner+v_id+".webm")
    
    val_v_link.append(temp)
```

```python
for v_ids in test_v_ids:
    temp = []
    
    for v_id in v_ids:
        
        try:
            l_idx = lines.index(line_beginner+v_id+".mp4")
            temp.append(line_beginner+v_id+".mp4")
                
        except:
            temp.append(line_beginner+v_id+".webm")
    
    test_v_link.append(temp)
```

### Save them for download purposes

```python
txt_save_path_train = "C:/Users/PC/Desktop/FewShotPhd/data/h100m_info/train/" 
txt_save_path_val = "C:/Users/PC/Desktop/FewShotPhd/data/h100m_info/val/"
txt_save_path_test = "C:/Users/PC/Desktop/FewShotPhd/data/h100m_info/test/"

# save train
for i in range(len(train_classes)):
    class_name = train_classes[i]
    class_name = "_".join(class_name.strip().split())
    
    with open(txt_save_path_train + str(class_name)+'.txt', 'w+') as f:
        for item in train_v_link[i]:
            f.write("%s\n" % item)

# save validation
for i in range(len(validation_classes)):
    class_name = validation_classes[i]
    class_name = "_".join(class_name.strip().split())
    
    with open(txt_save_path_val + str(class_name)+'.txt', 'w+') as f:
        for item in val_v_link[i]:
            f.write("%s\n" % item)
            
# save test
for i in range(len(test_classes)):
    class_name = test_classes[i]
    class_name = "_".join(class_name.strip().split())
    
    with open(txt_save_path_test + str(class_name)+'.txt', 'w+') as f:
        for item in test_v_link[i]:
            f.write("%s\n" % item)
```
