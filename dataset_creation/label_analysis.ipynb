```python
# Train file
train_df = pd.read_csv("C:/Users/PC/Desktop/FewShotPhd/data_hvu/hvu_classes/train_df.csv")
train_df.head(10)
```

```python
train_classes = list(set(train_df["action_name"].values.tolist()))
train_classes
```

```python
train_videos_path = "D:/HolisticVideoUnderstanding/train"
train_videos_classes = os.listdir(train_videos_path) 
train_video_ids = []

for t_class in train_videos_classes:
    full_path = os.path.join(train_videos_path, t_class)
    t_class_videos = os.listdir(full_path)
    
    for i in range(len(t_class_videos)):
        t_class_videos[i] = t_class_videos[i].split(".mp4")[0]
    
    train_video_ids.append(t_class_videos)
    
```

```python
# for each saved video check the label status for each semantic label category
train_analysis_dict = {}

for i in range(len(train_videos_classes)):
    action_class = train_videos_classes[i]
    
    train_analysis_dict[action_class] = {"id":[], "action":[], "event":[], "concept":[], "object":[], "attribute":[], "scene":[]};
    
    for id in train_video_ids[i]:
        
        train_analysis_dict[action_class]['id'].append(id)
        # split the tags of given video
        video_tag = train_df[train_df["youtube_id"] == id].Tags.values.tolist()[0].split("|")
        
        v_action = []
        v_event = []
        v_concept = []
        v_object = []
        v_attribute = []
        v_scene = []
        
       
        for tag in video_tag:
            if tag in action:
                v_action.append(tag)
            if tag in event:
                v_event.append(tag)
            if tag in concept:
                v_concept.append(tag)
            if tag in object:
                v_object.append(tag)
            if tag in attribute:
                v_attribute.append(tag)
            if tag in scene:
                v_scene.append(tag)
        
        train_analysis_dict[action_class]["action"].append(v_action)
        train_analysis_dict[action_class]["event"].append(v_event)
        train_analysis_dict[action_class]["concept"].append(v_concept)
        train_analysis_dict[action_class]["object"].append(v_object)
        train_analysis_dict[action_class]["attribute"].append(v_attribute)
        train_analysis_dict[action_class]["scene"].append(v_scene)     
        
        
        
        
```
