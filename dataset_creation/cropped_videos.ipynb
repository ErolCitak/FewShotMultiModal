```python
section = "train"
save_section = section+"_cut"
section_df = pd.read_csv("C:/Users/PC/Desktop/FewShotPhd/data_hvu/hvu_classes/"+section+"_df.csv")

# raw class videolari ve siniflar
raw_videos_path = "D:/HolisticVideoUnderstanding"
videos_path = os.path.join(raw_videos_path, section)
classes = os.listdir(videos_path)

save_path = os.path.join(raw_videos_path,save_section)

counter = 0
for video_cls in tqdm(classes):
    print("Now --> {}, {}".format(video_cls, counter))
    
    # create folder if not exists in target cut ds
    save_folder = os.path.join(save_path, video_cls)
    print(save_folder)
    
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)
    else:
        continue
               
    # full time videos
    videos = glob.glob(os.path.join(os.path.join(videos_path,video_cls), "*.mp4"))
    videos_id = []
    
    # construct id list
    for video in videos:
        video_path = video
        video_id = video.split("\\")[-1].split(".mp4")[0]
        videos_id.append(video_id)
    
    action_specific_df = section_df[section_df["action_name"] == video_cls]
    
    # ilgili sinifa ait dataframe i row by row gez!...
    for i in range(len(action_specific_df)):
        df_id = action_specific_df.iloc[i,1]
        
        if df_id in tqdm(videos_id):
            df_start = float(action_specific_df.iloc[i,2])
            df_end = float(action_specific_df.iloc[i,3])
            df_general_idx = int(action_specific_df.iloc[i,5])
            
            print(df_start, df_end, df_general_idx)
            
            try:
                ffmpeg_extract_subclip(os.path.join(os.path.join(videos_path, video_cls),df_id+".mp4"), df_start-0.50, df_end+0.50, targetname=save_folder+"/"+df_id+"_"+str(df_general_idx)+".mp4")
            except:
                pass
```
